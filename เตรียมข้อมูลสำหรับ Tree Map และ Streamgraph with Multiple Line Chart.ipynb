# เตรียมข้อมูลสำหรับ Tree Map

```python
import pandas as pd
converTable = dict()
converTable['facebook_id'] = str
converTable['twitter_id'] = str
converTable['instagram_id'] = str
converTable['account_id'] = str
accmap_df = pd.read_csv('accmap_4platform.csv', dtype = converTable)
accmap1_df = pd.read_csv('Account_Mapping_Cat_FB.csv', dtype = converTable)
fb_data = pd.read_csv('chordFB.csv', dtype = converTable)
yt_data = pd.read_csv('chordYT.csv', dtype = converTable)
ig_data = pd.read_csv('chordIG.csv', dtype = converTable)
twitter_data = pd.read_csv('chordTwitter.csv', dtype = converTable)
```

```python
fb = pd.merge(left = fb_data[['account_id','engagement']],
              right = accmap1_df[['account_name','facebook_id','type']],
              how = 'left',
              left_on = 'account_id',
              right_on ='facebook_id'
             )
fb['engagement_fb'] = fb['engagement'] 
fb = fb[fb['engagement_fb'].notna()].sort_values(by='engagement_fb', ascending = False)
fb = fb[fb['account_name'].notna()]
fb = fb[fb['engagement_fb'] > 0 ].sort_values(by='engagement_fb', ascending = False)
fb = fb.drop(columns = ['account_id','engagement'])
fb.to_csv('Treemap_FB.csv')
fb
```

```python
yt = pd.merge(left = yt_data[['account_id','engagement']],
              right = accmap1_df[['account_name','youtube_id','type']],
              how = 'left',
              left_on = 'account_id',
              right_on ='youtube_id'
             )
yt['engagement_yt'] = yt['engagement'] 
yt = yt[yt['engagement_yt'].notna()].sort_values(by='engagement_yt', ascending = False)
yt = yt[yt['account_name'].notna()]
yt = yt[yt['engagement_yt'] > 0 ].sort_values(by='engagement_yt', ascending = False)
yt = yt.drop(columns = ['account_id','engagement'])
yt.to_csv('Treemap_YT.csv')
yt
```

```python
ig = pd.merge(left = ig_data[['account_id','engagement']],
              right = accmap1_df[['account_name','instagram_id','type']],
              how = 'left',
              left_on = 'account_id',
              right_on ='instagram_id'
             )
ig['engagement_ig'] = ig['engagement'] 
ig = ig[ig['engagement_ig'].notna()].sort_values(by='engagement_ig', ascending = False)
ig = ig[ig['account_name'].notna()]
ig = ig[ig['type'].notna()]
ig = ig[ig['engagement_ig'] > 0 ].sort_values(by='engagement_ig', ascending = False)
ig = ig.drop(columns = ['account_id','engagement'])
ig.to_csv('Treemap_IG.csv')
ig
```

```python
twitter = pd.merge(left = twitter_data[['account_id','engagement']],
              right = accmap1_df[['account_name','twitter_id','type']],
              how = 'left',
              left_on = 'account_id',
              right_on ='twitter_id'
             )
twitter['engagement_twitter'] = twitter['engagement'] 
twitter = twitter[twitter['engagement_twitter'].notna()].sort_values(by='engagement_twitter', ascending = False)
twitter = twitter[twitter['account_name'].notna()]
twitter = twitter[twitter['engagement_twitter'] > 0 ].sort_values(by='engagement_twitter', ascending = False)
twitter = twitter.drop(columns = ['account_id','engagement'])
twitter.to_csv('Treemap_Twitter.csv')
twitter
```

# Streamgraph

```python
dateColumns = ['created_at']
converTable = dict()
converTable['account_id'] = str
converTable['facebook_id'] = str
fb_posttime = pd.read_csv('fb_posttime.csv', parse_dates=dateColumns,dtype = converTable)
fb_posttime.info()
```

```python
selectCols = ['account_id','account_display_name','created_at','engagement','comment','share','type','tag_friend','purchase_intention']
fb_posttime_area = fb_posttime[selectCols]

fillnaCols = ['engagement','comment','share','type','tag_friend','purchase_intention']
fb_posttime_area[fillnaCols] = fb_posttime_area[fillnaCols].fillna(0)
fb_posttime_area = fb_posttime_area[fb_posttime_area['type']!=0]
fb_posttime_area
```

```python
fb_posttime_area['year'] = pd.DatetimeIndex(fb_posttime_area['created_at']).year
fb_posttime_area['month'] = pd.DatetimeIndex(fb_posttime_area['created_at']).month
fb_posttime_area['day'] = pd.DatetimeIndex(fb_posttime_area['created_at']).day
fb_posttime_area['m_d_'] = fb_posttime_area['month'].astype(str).str.zfill(2) + fb_posttime_area['day'].astype(str).str.zfill(2)
# fb_posttime_area.to_csv('trial_areaplotinPW.csv')
fb_posttime_area
```

```python
fb_posttime_small = fb_posttime_area.groupby(['type','month','day']).sum()
fb_posttime_small = fb_posttime_small.reset_index()
fb_posttime_small['year'] = 2020
cols=["year","month","day"]
fb_posttime_small['date'] =fb_posttime_small[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
fb_posttime_small['date']=pd.to_datetime(fb_posttime_small['date'])
# fb_posttime_small.to_csv('trial_areaplotinPW_small.csv')
fb_posttime_small
```
